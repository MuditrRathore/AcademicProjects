```python
path = "C:\\Users\\hp\\Desktop\\NSE_dataset"
ayushi_files = glob.glob(path + "/ayushi*.csv")

ayushi_list = []

for filename in ayushi_files:
    df = pd.read_csv(filename)
    df = df.drop('Unnamed: 0',axis = 1)
    df = df.replace({ np.NaN : -1})
    ayushi_list.append(df)
    
print(ayushi_list)
```

```python
path = "C:\\Users\\hp\\Desktop\\NSE_dataset"
mudit_files = glob.glob(path + "/mudit*.csv")

mudit_list = []

for filename in ayushi_files:
    df = pd.read_csv(filename)
    df = df.drop('Unnamed: 0',axis = 1)
    df = df.replace({ np.NaN : -1})
    mudit_list.append(df)
    
print(len(mudit_list))
```
