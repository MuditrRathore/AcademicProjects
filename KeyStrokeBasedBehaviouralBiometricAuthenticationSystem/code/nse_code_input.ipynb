```python
#prompt = input("would you like to begin?")

first = time.time()
storing_list = []
release_list = []

def on_press(key):
    if Key.space is key:
        pass        # <-- Note: key.char
    elif Key.backspace is key:
        pass
    elif Key.enter is key:
        pass
    elif Key.shift is key:
        pass
    elif Key.space is not key:
        if key == Key.esc:
            # Stop listener
            return False
        a = key.char
        print(type(key))
        print('{0} pressed'.format(key))
        b = time.time() - first
        q = []
        q.append(a)
        q.append(b)
        storing_list.append(q)


def on_release(key):
    if Key.space is key:
        pass        # <-- Note: key.char
    elif Key.enter is key:
        pass
    elif Key.backspace is key:
        pass
    elif Key.shift is key:
        pass
    elif Key.space is not key:
        print('{0} release'.format(key))
        if key == Key.esc:
            # Stop listener
            return False
        print(time.time())
        w = []
        print(type(key))
    
        a = key.char
    
        b = time.time() - first
        w.append(a)
        w.append(b)
        release_list.append(w)

# Collect events until released
with Listener(
        on_press=on_press,
        on_release=on_release) as listener:
    listener.join()

print(storing_list)
```

```python
hash_store_list = []
hash_release_list = []

for i in range(256):
    hash_store_list.append([0])
    hash_release_list.append([0])
```

```python
for x in storing_list:
    hash_store_list[ord(str(x[0]))].append(x[1])
```

```python
for x in release_list:
    hash_release_list[ord(str(x[0]))].append(x[1])
```
